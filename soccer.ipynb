```python
df1 = pd.DataFrame(data_frame, columns=['long_name', 'wage_eur', 'value_eur'])
df1['difference'] = df1['value_eur'] - df1['wage_eur']
df1.sort_values('difference', ascending=False)
```

```python
import seaborn as sns
sns.set()

graph = sns.scatterplot(x='wage_eur', y='value_eur', data=df1)
graph
```

```python
from bokeh.plotting import figure,show
from bokeh.models import HoverTool

TOOLTIPS = HoverTool(tooltips=[
    ('index', '@index'),
    ('(wage_eur,value_eur)', '(@wage_eur, @value_eur)'),
    ('long_name', '@long_name')]
)

hover = HoverTool(tooltips=[])

p = figure(title='Soccer 2015', x_axis_label='wage_eur', y_axis_label='value_eur', plot_width=700, plot_height=700, tools=[TOOLTIPS])
p.circle('wage_eur', 'value_eur', size=10, source=df1)
show(p)
```
